```python
rho_0 = 0.05         
a = 10              

r = np.linspace(0.1, 100, 1000)

rho_r = rho_0 / (1 + (r/a)**2)**(2.5)
```

```python
plt.figure(figsize=(8,6))
plt.plot(r, rho_r, label='Plummer Profili')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('r [kpc]', fontsize=12)
plt.ylabel(r'$\rho(r)$ [yoğunluk birimi]', fontsize=12)
plt.title('Karanlık Madde - Plummer Yoğunluk Profili', fontsize=14)
plt.grid(True, which="both", ls=':')
plt.legend()
plt.show()
```

```python
Sigma0 = 500         
r0 = 40               

r = np.linspace(0, 100, 500)

Sigma_r = Sigma0 * np.exp(- (r**2) / (2 * r0**2))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.plot(r, Sigma_r, color='mediumslateblue', label='Gaz Yüzey Yoğunluğu')
plt.xlabel('r [kpc]', fontsize=12)
plt.ylabel(r'$\Sigma_{\rm gaz}(r)$ [M$_\odot$/pc$^2$]', fontsize=12)
plt.title('Gaz Yüzey Yoğunluğu Profili', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```

```python
Sigma_gas0 = 300     
r_gas = 15           
Sigma_star0 = 700    
r_star = 5           

r = np.linspace(0.1, 50, 400)

Sigma_gas = Sigma_gas0 * np.exp(-r / r_gas)
Sigma_star = Sigma_star0 * np.exp(-r / r_star)

Sigma_total = Sigma_gas + Sigma_star
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(r, Sigma_gas, label='Gaz Yüzey Yoğunluğu', color='deepskyblue')
plt.plot(r, Sigma_star, label='Yıldız Yüzey Yoğunluğu', color='goldenrod')
plt.plot(r, Sigma_total, label='Toplam Yüzey Yoğunluğu', color='green', linestyle='--')
plt.xlabel('r [kpc]', fontsize=12)
plt.ylabel(r'$\Sigma(r)$ [M$_\odot$/pc$^2$]', fontsize=12)
plt.title('Yüzey Yoğunluk Profili (Gaz, Yıldız, Toplam)', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

```python
M_gas_total = 3e10   # Total gas mass (M_sun)
r_g = 10             # Gas disk scale length (kpc)
M_star_total = 1e11  # Total stellar mass (M_sun)
r_d = 5              # Stellar disk scale length (kpc)

# Radius array (in kpc)
r = np.linspace(0.1, 50, 500)

# Gas mass profile (cumulative)
M_gas = M_gas_total * (1 - np.exp(-r / r_g) * (1 + r / r_g))

# Stellar mass profile (cumulative)
M_star = M_star_total * (1 - np.exp(-r / r_d) * (1 + r / r_d))

# Total baryonic mass profile
M_total = M_gas + M_star
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(r, M_gas, label='Gas Mass (Gaz Kütlesi)', color='deepskyblue')
plt.plot(r, M_star, label='Stellar Mass (Yıldız Kütlesi)', color='goldenrod')
plt.plot(r, M_total, label='Total Baryonic Mass (Toplam Barionik Kütle)', color='green')
plt.xlabel('Radius [kpc]', fontsize=12)
plt.ylabel('Mass [Solar Masses]', fontsize=12)
plt.title('Galaxy Baryonic Mass Profiles (Galaksi Barionik Kütle Profilleri)', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

```python
M_gas_total = 5e10  
r_gas = 15           
M_dust_total = 1e9   
r_dust = 5           
r = np.linspace(0.1, 50, 500)  

M_gas = M_gas_total * (1 - np.exp(-r / r_gas) * (1 + r / r_gas))
M_dust = M_dust_total * (1 - np.exp(-r / r_dust) * (1 + r / r_dust))

M_total = M_gas + M_dust
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(r, M_gas, label='Gas Mass (Gaz Kütlesi)', color='royalblue')
plt.plot(r, M_dust, label='Dust Mass (Toz Kütlesi)', color='peru')
plt.plot(r, M_total, label='Total Mass (Toplam Kütle)', color='seagreen')
plt.xlabel('Radius (kpc)')
plt.ylabel('Mass (Solar Masses)')
plt.title('Galaxy Gas and Dust Mass Profiles (Galaksi Gaz ve Toz Kütle Profilleri)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Function to print Fibonacci sequence
def fibonacci(n):
    a, b = 0, 1
    fib_sequence = []
    for _ in range(n):
        fib_sequence.append(a)
        a, b = b, a + b
    return fib_sequence

# Set the number of terms you want to print
num_terms = 10
print(fibonacci(num_terms))
```
